## matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotiing in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1, prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) #(width, height)

# 3. plot the data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("matplot-images/sample-plot.png")
```

### two options for subplots

```python
# option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
```

```python
#plot to each different axis
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(10000));
```

### which one should I use? (pyplot vs matplotlib OO Method)

* when plotting something quickly, OK to use pyplot method
* when plottin something more advanced, use the OO Method

```python
heart_disease.head()
```

```python
over_50 =  heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            ax=ax);

#ax.set_xlim([45,100]);
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Level",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add an horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# Add data ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Level",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add mean line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data ax1
scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

# Customize the plot
ax1.set(title="Heart Disease and Max heart rate Level",
      xlabel="Age",
      ylabel="Max heart rate");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add mean line
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');

```

```python
## Introduce a new style to the plot
plt.style.use('seaborn')

## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") # this change the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Level",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add an horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');

```

```python
## Introduce a new style to the plot
plt.style.use('seaborn-whitegrid')

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# Add data ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Level",
      xlabel="Age",
      ylabel="Cholesterol");

#change the X, Y axis of limits
ax0.set_xlim([50, 80])
#ax0.set_ylim([60, 200])


# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add mean line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data ax1
scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")

# Customize the plot
ax1.set(title="Heart Disease and Max heart rate Level",
      xlabel="Age",
      ylabel="Max heart rate");

#change the X, Y axis of limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add mean line
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');
```
